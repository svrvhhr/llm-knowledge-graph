```python
# Load model directly
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

tokenizer = AutoTokenizer.from_pretrained("Zual/propositioneur")
model = AutoModelForSeq2SeqLM.from_pretrained("Zual/propositioneur")
```

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

model_name = "google/flan-t5-large"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)
```

```python
text = "je suis Boualem MOKEDDEM, un étudiant à l'université de Paris-Saclay, je suis né en mai"

prompt = f"""
Tu es un système d’extraction des triplets de chaque phrase atomique

Tâche :
Extrais des triplets (SUJET, RELATION, OBJET) à partir du texte.

Règles :
- Chaque triplet doit représenter un fait explicite du texte.
- Utilise des relations normalisées en MAJUSCULES_AVEC_UNDERSCORE.
- Ne retourne PAS les mots "sujet", "RELATION" ou "objet".
- Ne retourne PAS le modèle de sortie.
- Ne devine rien.

Format de sortie STRICT :
[
  ["entité_1", "RELATION", "entité_2"]
]

Texte :
{text}
"""



inputs = tokenizer(
    prompt,
    return_tensors="pt",
    truncation=True
)

outputs = model.generate(
    **inputs,
    max_new_tokens=256,
    do_sample=False
)

result = tokenizer.decode(outputs[0], skip_special_tokens=True)
print(result)
```

```python
import spacy

# 1. Load the French model (ensure you have run: python -m spacy download fr_core_news_sm)
try:
    nlp = spacy.load("fr_core_news_sm")
except:
    import os
    os.system("python -m spacy download fr_core_news_sm")
    nlp = spacy.load("fr_core_news_sm")

def extract_french_triplets_robust(text):
    doc = nlp(text)
    triplets = []

    for token in doc:
        # We look for a verb as the root of the triplet
        if token.pos_ in ["VERB", "AUX"]:
            subj = ""
            obj = ""

            # Iterate through children to find Subject and Object
            for child in token.children:
                # French subjects: nsubj (standard) or nsubj:pass (passive)
                if "subj" in child.dep_:
                    # Get the full noun phrase for the subject (e.g., "Le petit chat")
                    subj = "".join([w.text_with_ws for w in child.subtree]).strip()

                # French objects: obj (direct), obl (indirect/prepositional), or attr (predicate)
                if child.dep_ in ["obj", "obl", "attr", "xcomp"]:
                    # Get the full noun phrase for the object
                    obj = "".join([w.text_with_ws for w in child.subtree]).strip()

            if subj and obj:
                triplets.append({
                    "subject": subj,
                    "relation": token.lemma_, # Using lemma (e.g., 'mangeait' -> 'manger')
                    "object": obj
                })

    return triplets

# --- TEST IT ---
test_sentences = [
    "Le système résout les équations.",
    "Marie donne un livre à Pierre.",
    "La solution est étudiée par les chercheurs."
]

for s in test_sentences:
    print(f"Phrase: {s}")
    print(f"Triplets: {extract_french_triplets_robust(s)}\n")
```

```python
import json
raw_text = "Marie a quitté Lyon en 2022.///Marie s'est installée à Berlin en 2022.///Marie travaille désormais comme ingénieure en intelligence artificielle à Berlin.///Marie a rejoint une start-up spécialisée dans la santé numérique après avoir terminé son doctorat.///Marie collabore régulièrement avec des chercheurs universitaires.///L'entreprise a levé des fonds importants l'année dernière.///Les fonds importants ont permis à l'entreprise d'ouvrir un nouveau bureau.///Les fonds importants ont permis à l'entreprise d'embaucher plusieurs employés."
sentences = [
    s.strip()
    for s in raw_text.split("///")
    if s.strip()
]

sentences = list(dict.fromkeys(sentences))

sentences_json = json.dumps(sentences, ensure_ascii=False)


prompt = f"""
You are a precise information extraction system.

The input is a JSON array of independent sentences.
Treat EACH sentence separately.

Your task:
Extract atomic (subject, relation, object) triples.

Rules:
1. Each triple represents ONE explicit fact.
2. Use normalized relation names in UPPER_SNAKE_CASE.
3. Subjects and objects must be concise noun phrases.
4. Do NOT invent facts.
5. Do NOT repeat identical triples.
6. Output ONLY valid JSON.

Output format (STRICT):
{{
  "triples": [
    {{
      "subject": "string",
      "relation": "UPPER_SNAKE_CASE",
      "object": "string"
    }}
  ]
}}

Input:
{sentences_json}
"""


raw_output = extractor(
    prompt,
    max_new_tokens=512,
    do_sample=False
)[0]["generated_text"]

print(raw_output)
```
